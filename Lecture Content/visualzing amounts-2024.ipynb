```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {'Movie title': ["Star Wars", "Pitch Perfect 3", "Greatest Showman", "Ferdinand" , "Jumanji"], 'Grossing': [71.57, 19.93, 8.81, 7.32, 36.17]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
fig = plt.figure()
ax= plt.axes()
ax.bar(df['Movie title'], df['Grossing']);
```

```python
fig = plt.figure(figsize=(10,5))
ax= plt.axes()
ax.bar (df['Movie title'], df['Grossing']);
```

```python
plt.style.use('ggplot')
fig = plt.figure(figsize=(15,10))
ax= plt.axes()
ax.bar(df['Movie title'], df['Grossing'])
ax.set_xlabel("Movie Title")
ax.set_ylabel(" Million Dollar");
```

```python
plt.style.use('ggplot')
fig = plt.figure(figsize=(10,5))
ax= plt.axes()
ax.bar (df['Movie title'], df['Grossing'], color= ['orange', 'purple', 'green', 'blue', 'cyan'])
ax.set_xlabel("Movie Title", fontsize= 12, color= 'navy')
ax.set_ylabel("Grossing (Million Dollar)", fontsize= 12, color= 'navy')
x_pos = np.arange(5)
ax.set_xticks(x_pos, df['Movie title'],  ha='center', rotation=0, fontsize=10, fontname='monospace');
```

```python
plt.style.use('ggplot')
fig = plt.figure(figsize=(10,10))
ax= plt.axes()
ax.barh (df['Movie title'], df['Grossing'], color= ['orange', 'purple', 'green', 'blue', 'cyan'])
#ax.set_ylabel("Movie Title", fontsize= 16, color= 'navy')
ax.set_xlabel(" Grossing (Million Dollar)", fontsize= 16, color= 'navy')
x_pos = np.arange(5)
ax.set_yticks(x_pos, df['Movie title'],  ha='right', rotation=0, fontsize=14, fontname='monospace');
```

let make a stacked and grouped bar chart to display counts by species and sex

```python
df2 = df.groupby(['species', 'sex'], as_index= False).size()
```

```python
df2
```

```python
male = df2[df2['sex'] =='male']
female = df2[df2['sex'] =='female']
```

```python
female
```

```python
x = np.arange(3)  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, male['size'], width, label='Male')
rects2 = ax.bar(x + width/2, female['size'], width, label='Female')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Count')
ax.set_title('Count by Species and Gender')
ax.set_xticks(x, male['species'])
ax.legend()


fig.tight_layout()

plt.show()
```

```python
width = 0.35      

fig, ax = plt.subplots()

ax.bar(male['species'], male['size'], width, label='Male')
ax.bar(female['species'], female['size'], width, male['size'],
       label='Female')

ax.set_ylabel('count')
ax.set_title('Count by Species and Gender')
ax.legend()

plt.show()
```
